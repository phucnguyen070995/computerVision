```
#HOMEWORK1
!gdown --id 1OivZGcmErX_1rd2gnLNE0N7SxiwnPHHC

import cv2
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import NoNorm

filename = './kyoto.jpg'
image = cv2.imread(filename, 0)
print(image.shape)
print(image)

# Show image
plt.figure(figsize=(12, 8))
plt.imshow(image, cmap='gray',norm=NoNorm())
plt.title('Original Image')
plt.axis('off')
plt.show()


#Tao array px(x)
image2 = image.flatten()
# print(image2)
print(image2.shape)
px = np.zeros(256)
# print(len(image2[image2 == 50]))
for i in range(256):
  px[i] = len(image2[image2 == i])
px /= image2.shape
# print(px)
print(max(px))
print(min(px))
print(sum(px))
```

```
cx = np.zeros(256)
cx[0] = px[0]
for i in range(1, 256):
  cx[i] = cx[i - 1] + px[i]
# print(cx)
col = range(256)
plt.bar(col,cx, color = 'blue')
cz = np.zeros(256)
for i in range(256):
  cz[i] = cz[i - 1] + 1/256
print(cz)
plt.bar(col,cz, color = 'green')
cimage = np.zeros(len(image2))
# print(cimage.shape)

for i in range(len(image2)):
  arr = np.full((256,), cx[image2[i]])
  sub_arr = abs(cz - arr)
  sub_arr = list(sub_arr)
  color = sub_arr.index(min(sub_arr))
  cimage[i] = color
cimage = np.reshape(cimage, (image.shape[0], image.shape[1]))
print(cimage.shape)
print(cimage)

# Show image
plt.figure(figsize=(12, 8))
plt.imshow(cimage, cmap='gray',norm=NoNorm())
plt.title('Original Image')
plt.axis('off')
plt.show()

```
