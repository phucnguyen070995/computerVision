```
import cv2
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import NoNorm

filename = './kyoto.jpg'
image = cv2.imread(filename, 1)
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
gray = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
cv2.imwrite('gray_kyoto.jpg', gray)
print(type(image))
print(image.shape)

plt.figure(figsize=(12, 8))
plt.imshow(image)
plt.title('Original Image')
plt.axis('off')
plt.show()

plt.figure(figsize=(12, 8))
plt.imshow(gray, cmap='gray', norm=NoNorm())
plt.title('Gray Image')
plt.axis('off')
plt.show()
```

```
filename = './kyoto.jpg'
image = cv2.imread(filename, 1)
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

redChannel = image.copy()
greenChannel = image.copy()
blueChannel = image.copy()
print(redChannel.shape)
redChannel[:,:,(1,2)] =0 
greenChannel[:,:,(0,2)] =0 
blueChannel[:,:,(0,1)] =0 

#Show image
plt.figure()
plt.imshow(redChannel)
plt.title('Red Channel')
plt.axis('off')
plt.show()

plt.figure()
plt.imshow(greenChannel)
plt.title('Green Channel')
plt.axis('off')
plt.show()

plt.figure()
plt.imshow(blueChannel)
plt.title('Blue Channel')
plt.axis('off')
plt.show()
```

```
#Load image
filename = './kyoto.jpg'
image = cv2.imread(filename, 0)
print(image.shape)
cimage = image.copy()


cimage = np.expand_dims(cimage,axis=-1)
cimage = np.repeat(cimage,3,axis=-1)
print(cimage.shape)

#Show image
plt.figure()
plt.imshow(image, cmap='gray',norm=NoNorm())
plt.title('Original Image')
plt.axis('off')
plt.show()

#Show image
plt.figure()
plt.imshow(cimage)
plt.title('Colorized of gray Image')
plt.axis('off')
plt.show()
```

```
#Load image
filename = './kyoto.jpg'
image = cv2.imread(filename, 0)
print(image.shape)
cimage = image.copy()


cimage = cimage.reshape(*cimage.shape, 1)
cimage = np.repeat(cimage,3,axis=-1)
print(cimage.shape)

#Show image
plt.figure()
plt.imshow(image, cmap='gray',norm=NoNorm())
plt.title('Original Image')
plt.axis('off')
plt.show()

#Show image
plt.figure()
plt.imshow(cimage)
plt.title('Colorized of gray Image')
plt.axis('off')
plt.show()
```

```
#Load image
filename = './kyoto.jpg'
image = cv2.imread(filename, 0)
print(image.shape)

cimage = np.repeat(image,3)
print(cimage.shape)
cimage = np.reshape(cimage, (image.shape[0], image.shape[1], 3))

#Show image
plt.figure()
plt.imshow(image, cmap='gray', norm=NoNorm())
plt.title('gray Image')
plt.axis('off')
plt.show()


plt.figure()
plt.imshow(cimage)
plt.title('Colorized Image')
plt.axis('off')
plt.show()
```

```
#Load image
filename = './kyoto.jpg'
image = cv2.imread(filename, 0)
print(image.shape)

cimage = np.tile(image.flatten(), 3)
print(cimage.shape)
#cimage = np.reshape(cimage, (image.shape[0], image.shape[1], 3))
cimage = np.reshape(cimage, ( 3, image.shape[0], image.shape[1]))
cimage = np.transpose(cimage, (1, 2,0))

print(cimage.shape)
#Show image
plt.figure()
plt.imshow(image, cmap='gray', norm=NoNorm())
plt.title('gray Image')
plt.axis('off')
plt.show()

plt.figure()
plt.imshow(cimage)
plt.title('colorized Image')
plt.axis('off')
plt.show()
```
