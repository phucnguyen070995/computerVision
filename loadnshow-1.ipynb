#Read images to gray format



```
import cv2
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import NoNorm


#Load image
filename = './kyoto.jpg'
image = cv2.imread(filename, 0)

#Show image
plt.figure()
plt.imshow(image, cmap='gray',norm=NoNorm())
plt.title('Original Image')
plt.axis('off')
plt.show()
```

#Read images to color format

```
#Load image
filename = './kyoto.jpg'
bgr_image = cv2.imread(filename, 1)
rgb_image = cv2.cvtColor(bgr_image, cv2.COLOR_BGR2RGB)
gray = cv2.cvtColor(rgb_image, cv2.COLOR_RGB2GRAY)

#Show image
plt.figure()
plt.imshow(rgb_image)
plt.title('Original Image')
plt.axis('off')
plt.show()

plt.figure()
plt.imshow(gray, cmap='gray',norm=NoNorm())
plt.title('Gray Image')
plt.axis('off')
plt.show()
```

#Save images

```
#Load image
filename = './kyoto.jpg'
bgr_image = cv2.imread(filename, 1)
rgb_image = cv2.cvtColor(bgr_image, cv2.COLOR_BGR2RGB)
gray = cv2.cvtColor(rgb_image, cv2.COLOR_RGB2GRAY)

#save to new image
cv2.imwrite('kyoto_gray.jpg', gray)

#load it again
image = cv2.imread(filename, 0)
plt.figure()
plt.imshow(gray, cmap='gray',norm=NoNorm())
plt.title('An Image')
plt.axis('off')
plt.show()
```
